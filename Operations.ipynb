## Read csv file and change headers name

```python
a = pd.read_csv(r'Desktop\new1.csv',names=['S.No','POSITION','DOB','FLAG','GENDER','DOJ'],header=0)
a
```

```python
a['NOW'] = datetime.today()  #Use to get today's date and time
a['DOB'] = to_datetime(a['DOB'])  #Create DOB column and arrange its date
b = (a['NOW']-a['DOB'])/timedelta64(1,'Y')  #Use to calculate age
a['Age'] = round(b)  #Use to create Age column and calculate round values 
a['DOJ'] = to_datetime(a['DOJ'])  #Create DOJ column and arrange its date
c = (a['NOW']-a['DOJ'])/timedelta64(1,'Y')  #Use to calculate Experience  
a['Experience'] = round(c)  #Use to create Experience column and calculate round values 
a['Chief'] = a[a['Experience']>=10] ##Use to create Chief column and check its seneority
a
a.drop(['FLAG','GENDER'],axis=1)  #Remove FLAG and GENDER column from table
```
